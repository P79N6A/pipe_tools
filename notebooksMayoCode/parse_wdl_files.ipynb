```python
""" update the branches dataframe files 
    Get the list of WDL files into dataframes
"""
mayo_clone_branches = ['dev','MayomicsVC', 'optional-trimseq', 'runtime-memory', 'workflow-generator']
base_dir = '../../../ncsa_genomics'
dataframes_output_directory = '../data/Mayomics_DataFrames'

for branch_used in mayo_clone_branches:
    may_dir = os.path.join(base_dir, branch_used)
    if not os.path.isdir(may_dir): continue
    print('processing git branch:\n', may_dir)

    mdl_dict = {}
    for dir_name, dir_list, files_list in os.walk(may_dir):

        if len(files_list) > 0 and not 'archives' in dir_name:
            for file_name in files_list:
                if file_name[-4:] == '.wdl':
                    mdl_dict[file_name] = os.path.join(dir_name, file_name)

    wdl_df = pd.DataFrame.from_dict(mdl_dict,orient='index')

    wdl_df.columns=['fullpathname']
    wdl_df.index.name = 'WDL file'

    from collections import OrderedDict
    srch_dict = OrderedDict([('n_workflows','workflow'), 
                            ('n_tasks','task'), 
                            ('n_inputs','input'), 
                            ('n_outputs','output'),
                            ('n_commands', 'command')])

    for col_name, _ in srch_dict.items():
        wdl_df[col_name] = 0

    for name, row in wdl_df.iterrows():
        lines = []
        try:
            with open(row.fullpathname, 'r') as fh:
                lines = fh.readlines()
        except:
            print(n, row.fullpathname)

        if len(lines) > 0:
            for l in lines:
                for col_name, key_word in srch_dict.items():
                    if key_word in l:
                        wdl_df[col_name].loc[name] += 1
    
    df_file_name = 'wdl_' + branch_used + '.tsv'
    df_file_name = os.path.join(dataframes_output_directory, df_file_name)
    wdl_df.to_csv(df_file_name, sep='\t')
    print(df_file_name, 'Writte')
```
